```python
df=movie_final.iloc[:,:6]
```

```python
movie=df.T.rename(columns=df.T.iloc[0]).drop(df.T.index[0])
```

```python
df['emotion1']=0
df['emotion2']=0
```

```python
for i ,name in enumerate(movie):
    df['emotion1'][i]=movie[name].sort_values(ascending=False).index[0]
    df['emotion2'][i]=movie[name].sort_values(ascending=False).index[1]
```

```python
df=df[['name','emotion1','emotion2']]
```

```python
movie_emotion2=movie_emotion[['name','emotion1','emotion2']]
```

```python
data=df.merge(movie_emotion2,on='name')
```

```python
data[data['name']=='덕혜옹주']
```

```python
# 주된 감정을 맞춘경우 
score=0
for name,a,b in zip(data['name'],data['emotion1_x'], data['emotion1_y']):
    if a!=b:
        score+=1
print(score)
```

```python
# 보조 감정을 맞춘경우
score2=0
for a,b in zip(data['emotion2_x'], data['emotion2_y']):
    if a==b:
        score2+=1
score2
```

```python
# 주 감정과 보조감정 모두를 고려한 score
score=0
for x1,x2,y1,y2 in zip(data['emotion1_x'],data['emotion2_x'],data['emotion1_y'],data['emotion2_y'] ):
    if x1==y1:
        score+=0.8
    elif (x1==y2 and x2==y1) or (x1==y1 and x1==y1):
        score+=1
    elif (x2==y2):
        score+=0.2
```
