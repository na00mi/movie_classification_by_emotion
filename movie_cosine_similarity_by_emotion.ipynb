```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import itertools
```

## 감정 정보만을 이용한 유사도 비교

```python
df=pd.read_csv('final_groupby_movie.csv', encoding='utf-8')
```

```python
len(df)
```

```python
df=df.iloc[:, 1:]
```

```python
df
```

```python
movie_list=list(df['영화제목'])
```

```python
features=df.iloc[:,1:].to_numpy()
```

```python
comb_ls=list(itertools.combinations(movie_list,2))
```

```python
len(comb_ls)
```

```python
movie_feat={name : features for name, features in zip(movie_list, features)}
```

```python
all_similarity=[]
for name in comb_ls:
    similarity=cosine_similarity([movie_feat[name[0]]],[movie_feat[name[1]]])
    all_similarity.append([name[0], name[1], similarity])
```

```python
emo_sim_df=pd.DataFrame(all_similarity, columns=['A','B', 'similarity'])
emo_sim_df.similarity=emo_sim_df.similarity.apply(lambda x:x[0][0])
emo_sim_df=emo_sim_df.sort_values('similarity', ascending=False)
```

```python
emo_sim_df[:50]
```

## 장르포함 유사도 비교

```python
genre=pd.read_csv('moive_final.csv', encoding='utf-8')
```

```python
movie_list=list(genre['name'])
```

```python
features=genre.iloc[:,1:].to_numpy()
```

```python
#이건 이제 모든 감정 다 포함(threshold 적용X)
```

```python
n_features=[]
for i in features:
    temp=[]
    for j in range(len(i)):
        if j==5 or j==6:
            pass
        else:
            temp.append(i[j])
    n_features.append(np.array(temp))
```

```python
n_features[0]
```

```python
comb_ls=list(itertools.combinations(movie_list,2))
movie_feat={name : features for name, features in zip(movie_list, n_features)}
```

```python
all_similarity=[]
for name in comb_ls:
    similarity=cosine_similarity([movie_feat[name[0]]],[movie_feat[name[1]]])
    all_similarity.append([name[0], name[1], similarity])
```

```python
sim_df=pd.DataFrame(all_similarity, columns=['A','B', 'similarity'])
sim_df.similarity=sim_df.similarity.apply(lambda x:x[0][0])
sim_df=sim_df.sort_values('similarity', ascending=False)
```

## threshold 0.2 설정해서 해보기

```python
th1_features=[]
for i in features:
    temp=[]
    for j in range(len(i)):
        if j==5 or j==6:
            pass
        else:
            if i[j]<0.2:
                temp.append(0)
            else:
                temp.append(i[j])
    th1_features.append(np.array(temp))
```
